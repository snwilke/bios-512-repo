### Question 5

Download the patient_names.csv and patient_properties.csv files from Canvas and read them into R. Manually set the date columns to be date variables. Print the first 10 observations of each.

````r
patient_names <- read.csv("patient_names.csv", stringsAsFactors = FALSE)
patient_properties <- read.csv("patient_properties.csv", stringsAsFactors = FALSE)

colnames(patient_names)
colnames(patient_properties)

patient_names$BIRTHDATE <- as.Date(patient_names$BIRTHDATE, format = "%m/%d/%Y")

head(patient_names, 10)
head(patient_properties, 10)
````

### Question 6
In the data frame pulled from patient_properties, you'll notice that the data is long, not wide. Do a pivot to make the properties their own columns. Print the first 10 observations after you do so.

````r
library(tidyverse)

patient_properties_wide <- patient_properties %>%
  pivot_wider(
    id_cols = ID,            
    names_from = property,   
    values_from = value
  )
head(patient_properties_wide, 10)
````

### Question 7
Perform a left join of the names and properties_wide data frames by the ID column and print the first 10 rows.

````r
library(dplyr)

patient_names_and_propwide <- patient_names %>%
  left_join(patient_properties_wide, by = "ID")
head(patient_names_and_propwide, 10)
````

### Question 8
Notice something interesting about the names in our data set. Fix the name formatting and print the first 10 observations.

````r
patient_names_and_propwide$FIRST <- gsub("[0-9]", "", patient_names_and_propwide$FIRST)
patient_names_and_propwide$LAST  <- gsub("[0-9]", "", patient_names_and_propwide$LAST)

library(stringr)

patient_names_and_propwide$FIRST <- str_to_title(patient_names_and_propwide$FIRST)
patient_names_and_propwide$LAST  <- str_to_title(patient_names_and_propwide$LAST)

head(patient_names_and_propwide, 10)
````

### Question 9
Using a for statement to loop through the categorical variables (excluding name and ID), print the counts of each unique value in descending order, using the mdpre() function for formatting.

````r
categorical_cols <- names(patient_names_and_propwide)[
  sapply(patient_names_and_propwide, is.factor) | sapply(patient_names_and_propwide, is.character)
]
categorical_cols <- setdiff(categorical_cols, c("ID", "NAME"))

for (col in categorical_cols) {
  cat("\nCounts for", col, ":\n")
  counts <- table(patient_names_and_propwide[[col]])
  counts <- sort(counts, decreasing = TRUE)
  print(mdpre(counts))
}
````

### Question 10
If you see any weird values, get rid of the ones that don't make sense, and combine the ones that are formatted wrong. Don't forget ot check the dates! Print the new tables for categorical values, and print the date ranges.

````r
# Standardize and clean categorical variables
patient_names_and_propwide$GENDER <- toupper(patient_names_and_propwide$GENDER)
patient_names_and_propwide$GENDER[patient_names_and_propwide$GENDER %in% c("FEMALE", "F")] <- "F"
patient_names_and_propwide$GENDER[patient_names_and_propwide$GENDER %in% c("MALE", "M")] <- "M"

patient_names_and_propwide$ETHNICITY <- tolower(patient_names_and_propwide$ETHNICITY)
patient_names_and_propwide$ETHNICITY[patient_names_and_propwide$ETHNICITY %in% c("nonhispani", "nonhispanic")] <- "nonhispanic"
patient_names_and_propwide$ETHNICITY[patient_names_and_propwide$ETHNICITY %in% c("hispani", "hispanic")] <- "hispanic"

patient_names_and_propwide$RACE <- tolower(patient_names_and_propwide$RACE)
patient_names_and_propwide$RACE[patient_names_and_propwide$RACE %in% c("asiann")] <- "asian"
patient_names_and_propwide$RACE[patient_names_and_propwide$RACE %in% c("hawaiian")] <- "other"  # assuming grouping

patient_names_and_propwide$MARITAL <- toupper(patient_names_and_propwide$MARITAL)
patient_names_and_propwide$MARITAL[patient_names_and_propwide$MARITAL %in% c("M", "MALE")] <- "M"
patient_names_and_propwide$MARITAL[patient_names_and_propwide$MARITAL %in% c("S", "FINE")] <- "S"

# Remove invalid or future dates
patient_names_and_propwide$BIRTHDATE <- as.Date(patient_names_and_propwide$BIRTHDATE, format = "%y-%m-%d")

patient_names_and_propwide <- patient_names_and_propwide[!is.na(patient_names_and_propwide$BIRTHDATE) & patient_names_and_propwide$BIRTHDATE <= Sys.Date(), ]

# Print cleaned tables
cat("Counts for GENDER:\n")
print(table(patient_names_and_propwide$GENDER))

cat("\nCounts for ETHNICITY:\n")
print(table(patient_names_and_propwide$ETHNICITY))

cat("\nCounts for RACE:\n")
print(table(patient_names_and_propwide$RACE))

cat("\nCounts for MARITAL:\n")
print(table(patient_names_and_propwide$MARITAL))

cat("\nDate range:\n")
print(range(patient_names_and_propwide$BIRTHDATE))
````

### Question 11
Make a histogram of the ages of patients by gender. 

````r
patient_names_and_propwide <- patient_names_and_propwide %>%
  mutate(AGE = as.numeric(difftime(Sys.Date(), BIRTHDATE, units = "days")) %/% 365)

# Make histogram of age counts, separated by gender
ggmd(
  ggplot(patient_names_and_propwide, aes(x = AGE, fill = GENDER)) +
    geom_histogram(binwidth = 5, position = "dodge", color = "black") +
    labs(
      title = "Distribution of Patient Ages by Gender",
      x = "Age (years)",
      y = "Count",
      fill = "Gender"
    ) +
    theme_minimal() +
    theme(
      axis.text.x = element_text(angle = 45, hjust = 1),
      text = element_text(size = 12)
    )
)
````
